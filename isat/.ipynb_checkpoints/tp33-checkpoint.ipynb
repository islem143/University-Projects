```python
import matplotlib.pyplot as plt
import numpy as np
import math as mt
from sklearn.metrics import mean_squared_error
from PIL import Image



plt.rc("font", size=20)
img = np.asarray(plt.imread("./img.jpeg"), dtype="float32")
img /= 255.0

plt.figure(figsize=(32, 16))
plt.imshow(img)
plt.axis('off')
```

```python
Hx = [
    [1,0,-1],
    [2,0,-2],
    [1,0,-1]]
Hy = [
    [1,2,1],
    [0,0,0],
    [-1,-2,-1]
]

def filtreSobel(img, maskx=Hx,masky=Hy,seuil=1):
    res = np.ones((len(img),len(img[0])))
    ii = len(maskx)//2
    jj = len(maskx[0])//2
    for i in range(ii,len(img)-ii):
         for j in range(jj,len(img[i])-jj):
                fx = np.mean(np.multiply(img[i-ii:i+ii+1,j-jj:j+jj+1], maskx))
                fy = np.mean(np.multiply(img[i-ii:i+ii+1,j-jj:j+jj+1], masky))
                r = mt.sqrt(fx**2+fy**2)
                if(r<seuil):
                    res[i][j]= 0
    return res
```

```python
plt.figure(figsize=(32, 16))
for i in range(3):
    plt.subplot(1,3,i+1)
    plt.title(f'La placien ({["Rouge","Vert", "Bleu"][i]})')
    plt.axis('off')
    plt.imshow(filtreSobel(img[:,:,i], seuil=[.1,.1,.1][i]), cmap="gray")
```

```python
mask =  [
    [0,-1,0],
    [-1,8,-1],
    [0,-1,0]]

def filtreLaPla(img, mask=mask, seuil=1.5):
    res = np.zeros((len(img),len(img[0])))
    ii = len(mask)//2
    jj = len(mask[0])//2
    for i in range(1,len(img)-ii):
         for j in range(1,len(img[i])-jj):
                r= (np.mean(np.multiply(img[i-ii:i+ii+1,j-jj:j+jj+1], mask)))
                if(r<seuil):
                    res[i][j]= 1
                
    return res
```

```python
plt.figure(figsize=(32, 16))
for i in range(3):
    plt.subplot(1,3,i+1)
    plt.title(f'La placien ({["Rouge","Vert", "Bleu"][i]})')
    plt.axis('off')
    plt.imshow(filtreLaPla(img[:,:,i], seuil=[.01,.05,.05][i]), cmap="gray")
```

```python
def otsu(_img, is_normalized=True):
    image = _img.copy()
    bins_num = 256
 
    # Get the image histogram
    hist, bin_edges = np.histogram(image, bins=bins_num)
    
    # Get normalized histogram if it is required
    if is_normalized:
        hist = np.divide(hist.ravel(), hist.max())
    
    # Calculate centers of bins
    bin_mids = (bin_edges[:-1] + bin_edges[1:]) / 2.
    
    # Iterate over all thresholds (indices) and get the probabilities w1(t), w2(t)
    weight1 = np.cumsum(hist)
    weight2 = np.cumsum(hist[::-1])[::-1]
    
    # Get the class means mu0(t)
    mean1 = np.cumsum(hist * bin_mids) / weight1
    # Get the class means mu1(t)
    mean2 = (np.cumsum((hist * bin_mids)[::-1]) / weight2[::-1])[::-1]
    
    inter_class_variance = weight1[:-1] * weight2[1:] * (mean1[:-1] - mean2[1:]) ** 2
    
    # Maximize the inter_class_variance function val
    index_of_max_val = np.argmax(inter_class_variance)
    
    threshold = bin_mids[:-1][index_of_max_val]
    binary_mask = image > threshold
    selection = image.copy()
    selection[~binary_mask] = 0
    return selection

```

```python
plt.figure(figsize=(32, 16))
plt.title(f'Segmentation par OTSU')
plt.axis('off')
plt.imshow(otsu(img), cmap="gray")
plt.figure(figsize=(32, 16))
for i in range(3):
    plt.subplot(1,3,i+1)
    plt.title(f'Segmentation par OTSU ({["Rouge","Vert", "Bleu"][i]})')
    plt.axis('off')
    plt.imshow(otsu(img[:,:,i]), cmap="gray")
```
